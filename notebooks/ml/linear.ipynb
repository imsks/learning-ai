```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Data
data = {
    "area": [1000, 1200, 1500, 1800, 2000],
    "price": [40, 50, 60, 70, 80]
}
df = pd.DataFrame(data)
```

```python
# Step 2: Model
model = LinearRegression()
model.fit(df[['area']], df['price'])
```

```python
# Predict
predicted_price = model.predict([[1300]])
predicted_price
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Create some sample data for a simple linear relationship
# Let's say house size (in sq ft) is our input and price (in thousands USD) is our output
np.random.seed(0)
house_sizes = np.random.rand(50, 1) * 2000 + 1000 # 50 houses, sizes from 1000 to 3000 sq ft
prices = 50 * house_sizes + np.random.randn(50, 1) * 200 + 50 # Price = 50*size + some random noise

# 2. Create and train (fit) the Linear Regression model
model = LinearRegression()
model.fit(house_sizes, prices)

# 3. Get the model's parameters
slope = model.coef_[0][0]
intercept = model.intercept_[0]
print(f"Model's slope (coefficient): {slope:.2f}")
print(f"Model's intercept: {intercept:.2f}")

# 4. Make a prediction for a new house
new_house_size = np.array([[2000]])
predicted_price = model.predict(new_house_size)
print(f"\nPredicted price for a 2000 sq ft house: ${predicted_price[0][0]:.2f}K")

# 5. Visualize the data and the regression line
plt.scatter(house_sizes, prices, color='blue', label='Actual House Prices')
plt.plot(house_sizes, model.predict(house_sizes), color='red', linewidth=3, label='Linear Regression Line')
plt.title('Linear Regression for House Price Prediction')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (thousands USD)')
plt.legend()
plt.grid(True)
plt.show()
```
