```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate a range of numbers from -10 to 10
z_values = np.linspace(-10, 10, 100)
probabilities = sigmoid(z_values)

# Plot the Sigmoid curve
plt.figure(figsize=(8, 6))
plt.plot(z_values, probabilities, color='red')
plt.title('The Sigmoid (Logistic) Function')
plt.xlabel('z (any real number)')
plt.ylabel('Probability (between 0 and 1)')
plt.axhline(0.5, color='black', linestyle='--', linewidth=0.8)
plt.show()
```
