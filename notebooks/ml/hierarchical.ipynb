```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

X = np.array([[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]])

linked = linkage(X, 'single')

dendrogram(linked, orientation='top', labels=["lama", "cow", "elephant", "monkey", "rabbit", "shark"], distance_sort='descending', show_leaf_counts=True)
plt.show()
```
